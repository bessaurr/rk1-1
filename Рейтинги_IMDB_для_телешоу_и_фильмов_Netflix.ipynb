## 1. Введение: Рейтинги IMDB для телешоу и фильмов Netflix

Набор данных Netflix TV Shows and Movies предоставляет исчерпывающую информацию о различных названиях, доступных на популярной потоковой платформе. Набор данных включает в себя такие детали, как название названия, его тип (телешоу или фильм), краткое описание контента, год его выпуска, возрастной рейтинг, продолжительность показа (для телешоу: продолжительность эпизодов, для фильмов: продолжительность), оценка IMDb и голоса IMDb.

Этот набор данных получен отсюда: https://www.kaggle.com/datasets/thedevastator/netflix-imdb-scores/data

Анализируя этот набор данных, мы можем получить представление о распределении оценок и рейтингов IMDb как телешоу, так и фильмов, доступных на Netflix. Эта информация может помочь нам понять популярность и восприятие названий на основе оценок пользователей.

## 2. Описание данных



Набор данных был тщательно отобран для обеспечения точности и актуальности. Он исключает любые нулевые значения из оценок IMDb для обеспечения целостности данных. Каждая запись также содержит идентификатор, соответствующий JustWatch (платформе для легального потокового вещания), а также соответствующий идентификатор заголовка на IMDb.

title: Название телешоу или фильма.

В этом столбце содержатся названия различных телешоу и фильмов, доступных на Netflix. Вы можете использовать эту информацию для идентификации конкретных заголовков в наборе данных.

type: указывает, является ли запись телешоу или фильмом.

Столбец типа классифицирует каждую запись как телешоу или фильм. Вы можете отфильтровать свой анализ на основе этих категорий, чтобы сосредоточиться либо на телешоу, либо на фильмах.

describe: Краткое описание телешоу или фильма.

Столбец описания содержит краткое описание сюжета или сюжетной линии каждого телешоу или фильма. Эта информация поможет вам получить общее представление о каждом заголовке, прежде чем приступать к дальнейшему анализу.

release_year: год выхода телешоу или фильма.

В этом столбце указан год выпуска каждого названия в числовом формате. Вы можете использовать эти данные для изучения тенденций с течением времени, группируя и агрегируя названия по годам их выпуска.

age_certification: возрастной рейтинг телешоу или фильма.

В столбце age_certification указаны возрастные рейтинги, присвоенные каждому заголовку, указывающие, подходят ли они для общей аудитории (например, всех возрастов) или ограничены из-за контента для взрослых (например, с рейтингом R). Анализ этого атрибута позволяет понять, какой тип контента Netflix предлагает на разных возрастных уровнях.

runtime: продолжительность эпизодов телешоу ИЛИ продолжительность фильмов.

В столбце времени выполнения указана продолжительность эпизодов телешоу или продолжительность фильмов в числовом формате. Эта информация может помочь вам определить более короткие или длинные заголовки в зависимости от времени их выполнения и сравнить их в ходе анализа.

imdb_score: оценка телешоу или фильма на IMDB.

В этом столбце отображается оценка IMDB, присвоенная каждому заголовку, отражающая его общее качество и популярность на IMDB. Используйте этот показатель для оценки и ранжирования различных изданий на основе их рейтингов, потенциально раскрывая интересные закономерности или идеи.

imdb_votes: количество голосов, полученных телешоу или фильмом.

## 4. Подготовка данных: очистка и формирование



### Reading the Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Netflix TV Shows and Movies.csv')
df = df.drop_duplicates()
```

```python
df = pd.read_csv('Netflix TV Shows and Movies.csv')
```

```python
df.head(10)
```

### Обработка пропущенных значений:

```python
df = df.dropna()
```

### Проверка дубликатов

```python
df.duplicated().sum()
```

### Обработка отсутствующих значений:

```python
df['runtime'].fillna(df['runtime'].mean(), inplace=True)

df['age_certification'].fillna(df['age_certification'].mode()[0], inplace=True)
```

### Преобразование данных:

```python
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')

df['runtime'] = pd.to_numeric(df['runtime'], errors='coerce')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
df.tail(10)
```

```python
average_score_by_type = df.groupby('type')['imdb_score'].mean()
print("Средние оценки по типу контента:\n", average_score_by_type)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=average_score_by_type.index, y=average_score_by_type.values)
plt.title('Средние оценки по типу контента')
plt.xlabel('Тип контента')
plt.ylabel('Средняя оценка IMDb')
plt.show()
```

```python
correlation_runtime_score = df['runtime'].corr(df['imdb_score'])
print("Корреляция между продолжительностью и оценками IMDb:", correlation_runtime_score)
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='runtime', y='imdb_score', data=df)
plt.title('Влияние продолжительности на оценки IMDb')
plt.xlabel('Продолжительность (в минутах)')
plt.ylabel('Оценка IMDb')
plt.show()
```

```python
correlation_year_score = df['release_year'].corr(df['imdb_score'])
print("Корреляция между годом выпуска и оценками IMDb:", correlation_year_score)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='release_year', y='imdb_score', data=df)
plt.title('Корреляция между годом выпуска и оценками IMDb')
plt.xlabel('Год выпуска')
plt.ylabel('Оценка IMDb')
plt.show()
```

```python
average_score_by_year = df.groupby('release_year')['imdb_score'].mean()
print("Средние оценки по годам:\n", average_score_by_year)
```

```python
plt.figure(figsize=(12, 6))
average_score_by_year.plot(kind='line')
plt.title('Изменение средних оценок по годам')
plt.xlabel('Год выпуска')
plt.ylabel('Средняя оценка IMDb')
plt.show()
```

```python
average_score_by_age_rating = df.groupby('age_certification')['imdb_score'].mean()
print("Средние оценки по возрастному рейтингу:\n", average_score_by_age_rating)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=average_score_by_age_rating.index, y=average_score_by_age_rating.values)
plt.title('Средние оценки по возрастному рейтингу')
plt.xlabel('Возрастной рейтинг')
plt.ylabel('Средняя оценка IMDb')
plt.show()
```
